```python
import numpy as np
import pandas as pd

import mimesis as mim


from random  import randint
```

```python
def get_names(n:int):
    person = mim.Person('ru')

    for i in  range(n):
        yield person.full_name()

def get_boleans(n:int):
    for i in range(n):
        yield randint(0,1)

for i in get_names(10):
    print(i)
```

```python
count_player = 40


full_names = np.array([i for i in get_names(count_player)])
worker = np.array([i for i in  get_boleans(count_player)])

PLAYERS = pd.DataFrame(
    {
                    'Full_name':full_names, 
                   'Holiday_player':worker
    }
)
PLAYERS.to_csv('PLAYERS', header=False)
```

```python
Campaigns = np.array([
    'https://dungeonsanddragons.ru/modulzip/Hoard%20of%20the%20Dragon%20Queen%20RUS.pdf',
    'https://dungeonsanddragons.ru/modulzip/DDEX1-2%20-%20Secrets%20of%20Sokol%20Keep%20RUS.pdf',
    'https://dungeonsanddragons.ru/modulzip/DDEX1-11%20-%20Dark%20Pyramid%20of%20the%20Sorcerer%20Isle%20RUS.pdf',
    'https://dungeonsanddragons.ru/ADVENTURES/DDEX2-1%20-%20City%20of%20Danger%20RUS.pdf',
    'https://dungeonsanddragons.ru/ADVENTURES/DDEX2-5%20-%20Flames%20of%20Kythorn%20RUS.pdf',
    'https://dungeonsanddragons.ru/ADVENTURES/DDEX2-7_-_Bounty_in_the_Bog_RUS.pdf'
])

Campaigns_names = np.array([
    'Hoard of the Dragon Queen',
    'Secrets of Sokol Keep',
    'Dark Pyramid of Sorcerer’s Isle',
    'City of Danger',
    'Flames of Kythorn',
    'Bounty in the Bog',
])

is_done = [1, 1, 0, 0, 0, 0]

data = np.array([Campaigns_names, Campaigns, is_done]).T

Campaigns_data = pd.DataFrame(data, columns=['name', 'link', 'is_done'])

Campaigns_data
```

```python
Campaigns_data.to_csv('Campaigns', header=False)
```

```python
from mimesis import Generic as gnr

d = gnr('ru')

def get_dates(n:int):
    result = []
    for i in range(n):
        result.append(d.datetime.date())
    result.sort()
    return result
```

```python
id_campaign = []
game_date = []
session_num = []

for i in range(Campaigns_data.shape[0]):
    session_len = randint(5, 16)
    for k in range(session_len):
        session_num.append(k)
        id_campaign.append(i)
    game_date += get_dates(session_len)
```

```python

game_session = np.array([id_campaign, game_date, session_num]).T
SESSION = pd.DataFrame(game_session, columns = ['campaign_id', 'game_date', 'session_num'])
SESSION
```

```python
SESSION.to_csv('SESSION', header=False)
```
